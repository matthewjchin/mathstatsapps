```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
```

Birthday problem involves finding probability that at least one person's DOB matches another person's DOB in a group of $n$ people, where $n$ is the number of people in a group. Below is what computes the birthday paradox for any $n$ people included in a counting.  

```python
# Set up variables as if the birthday problem means that with one person there is one match
# Start probability there is at least one match at 100% or 1/1
# Also, there are 365 days in a year
prob_num = 1
prob_denom = 1
days = 365 
match_prob = 1

# Try different numbers until probability that at least more than one match hits 50%
# Use following numbers: 20, 23, 35, 40, 50
# Take these numbers and put them in a list
# Note if num_people[1] = 23, prob is 50.73% - at this many people the probability crosses 50%

num_people = [20,23,25,35,40,50]

# Create separate list that will include header for all data to be written to csv file
write_ppl = [None] * (len(num_people))
```

Go through each element in array. Find each probability of $n$ given people and the probability that at least one pair of two students minimum has a birthday match. Use the birthday paradox formula with respect to the number of people in a sample to find the chances that at least one match of two students exists. Take the number of students and its respective match probability and add to a new list to be written into a .csv file.  

```python
# Use for loop in range of number of students in classroom to compute birthday paradox 
# and the number of days in a calendar year. 
for a in range(len(num_people)):
    prob_num = 1
    prob_denom = 365**num_people[a] # 365 to power of number of students in sample
    days = 365 # for each number of students in list reset total days in year to 365
    match_prob = 1
    for b in range(num_people[a]):
        prob_num *= (days-b)
        match_prob = 1 - (prob_num / prob_denom)
    write_ppl[a] = [num_people[a], match_prob]
    # write the number of students and its chance that at least one match of two students true
```

```python
# Write all n students, bday match chances to csv file
file = "chance_pair_bday_matches.csv"
head = ['n students', 'chance of match']
with open(file, 'w') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(head)
    writer.writerows(write_ppl)
dobs = pd.read_csv(file)
```

```python
dobs['n students'] # return output for n students
```

```python
dobs["chance of match"] # return output for chances of at least one match for n students
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.bar(dobs["n students"], dobs["chance of match"])
ax.set_xlabel("Number of Students")
ax.set_ylabel("Chance of Birthday Match")
plt.show()
```

Here, if $n = 23$ people, the probability that there is at least one birthday match between two people is at least $50\%$ is approximately $50.73\%$. 

If $n = 35$, the probability that there is at least one birthday match between two people is approximately $81.4\%$.

If $n = 40$, the probability that there is at least one birthday match between two people is approximately $89.1\%$. 

If $n = 50$, the probability that there is at least one birthday match between two people is approximately $97\%$.

Write a program that takes a sample of a given size from an exponential distribution and computes the mean of the sample.

```python
# Set up exponential distribution of a set of randomly generated numbers of float type
# Size of set is 100 random numbers
exp_dist = np.random.exponential(scale=10, size=100) 

# Return the mean of the exponential distribution of the sample
m = exp_dist.mean()
print(m)
```

```python
# Create a histogram of the sample of exponential distribution
fig, ax = plt.subplots(figsize=(10,8))
ax.hist(exp_dist, bins=10, label="exp dist sample of 10") # take the distribution set
plt.show()
```

```python
# Create an empty 2D array with 6 rows and 1000 columns. 
# Each row represents k size of 1000 samples in of means for each of 1000 trials of k in sample.
# The k values: k = 1, 5, 10, 20, 30, 50. 

times = 1000
mean_data = np.empty([6,times])
k = [1,5,10,20,30,50]
```

```python
# Iterate through 2D array row and each of 1000 columns for each k value
# Nested for loop assurers that each row k generates 1000 means of k
# Means generated to be added to 2D array
# Generate a set of k random variables in exponential distribution  
# After generating k values take its mean and insert it to the 2D array
for count_1 in range(len(mean_data)):
    for count_2 in range(len(mean_data[count_1])):
        expdst = np.random.exponential(scale=1, size=k[count_1])
        np.insert(mean_data, k[count_1], expdst.mean())
```

```python
# The following are the six histograms for means of k exp dist observations 1000 times. 

# Set scale of histogram of exp dist of means
# Display all data on histogram for k = 1
fig, ax = plt.subplots(figsize=(10,8))
ax.hist(mean_data[0], label="mean exp dist k=1", bins=2)
plt.show()
```
